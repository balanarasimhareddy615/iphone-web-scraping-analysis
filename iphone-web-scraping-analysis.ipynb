```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

```python
url = "https://www.flipkart.com/search?q=iphone%2017&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
response.text
```

```python
html_text = response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_text)
soup.text
```

```python
soup.find("div",class_ = "hZ3P6w DeU9vF").text
```

```python
product_name = []
product_price = []
product_rom = []
product_display = []
product_camera = []
product_processor = []
product_warranty = []
product_rating = []
```

```python
price = soup.find_all("div",class_ = "hZ3P6w DeU9vF")
for i in price:
    product_price.append(re.sub(r"\D","",i.text))
```

```python
product_price
```

```python
product_name = []
```

```python
name = soup.find_all("div", class_ = "RG5Slk")
for i in name:
    product_name.append(i.text)
```

```python
product_name
```

```python
product_rating = []
```

```python
rating = soup.find_all("div", class_ = "MKiFS6")
for i in rating:
    product_rating.append(i.text)
```

```python
product_rating
```

```python
features = soup.find_all("li", class_ = "DTBslk")
for i in features:
    print(i.text)
```

```python
features = soup.find_all("li", class_ = "DTBslk")
for i in features:
    if "ROM" in i.text:
        product_rom.append(i.text)
    elif "Display" in i.text:
        product_display.append(i.text)
    elif "Camera" in i.text:
        product_camera.append(i.text)
    elif "Processor" in i.text:
        product_processor.append(i.text)
    elif "Warranty" in i.text:
        product_warranty.append(i.text)
```

```python
print("product_name",len(product_name))
print("product_price",len(product_price))
print("product_rom",len(product_rom))
print("product_display",len(product_display))
print("product_camera",len(product_camera))
print("product_processor",len(product_processor))
print("product_warranty",len(product_warranty))
print("product_rating",len(product_rating))
```

```python
correct_length = len(product_name)
product_rating = product_rating[:correct_length]
```

```python
print("product_rating",len(product_rating))
```

```python
import pandas as pd
products_df = pd.DataFrame({
    "product_name":product_name,
    "product_price":product_price,
    "product_rom":product_rom,
    "product_display":product_display,
    "product_camera":product_camera,
    "product_processor":product_processor,
    "product_warranty":product_warranty,
    "product_rating":product_rating})
```

```python
products_df
```

```python
products_df.info()
products_df.isna().sum()
products_df.duplicated().sum()
```

```python
products_df.drop_duplicates(inplace=True)
```

```python
products_df['product_price'] = products_df['product_price'].astype(int)
```

```python
products_df['rom_gb'] = (
    products_df['product_rom']
    .str.extract(r'(\d+)')
    .astype(int)
)
```

```python
products_df.groupby('rom_gb')['product_price'].mean()
```

```python
products_df['product_rating'].describe()
```

```python
import matplotlib.pyplot as plt

products_df['product_price'].plot(kind='hist', bins=5)
plt.title('Price Distribution of iPhone Models')
plt.xlabel('Price')
plt.show()
```

```python
products_df.groupby('rom_gb')['product_price'].mean().plot(kind='bar')
plt.title('Average Price by Storage')
plt.xlabel('Storage (GB)')
plt.ylabel('Price')
plt.show()
```
